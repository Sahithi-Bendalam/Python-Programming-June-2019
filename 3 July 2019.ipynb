```python
# Function to read csv data
filepath = 'DataFiles/AppleStore.csv'
import pandas as pd
def readCSVdata(filepath):
    return pd.read_csv(filepath)
readCSVdata(filepath)
```

```python
# List of all unique Prime_Genres(categories) in the dataset
incomedf =readCSVdata(filepath)
def primeGeners(fp):
    l=[]
    li=[]
    for i in range(len(fp.values[0::])):
        column = fp.values[i][12]
        l.append(column)
    for i in l:
        if i not in li:
            li.append(i)
    print(li)
    return 
primeGeners(incomedf)
        
```

```python
# Category with highest number of apps
incomedf =readCSVdata(filepath)
def HighestNoOfApps(fp):
    l=[]
    d={}
    li=[]
    for i in range(len(fp.values[0::])):
        column = fp.values[i][12]
        l.append(column)
        if l[i] not in d:
            d[l[i]]=1
        else:
            d[l[i]]+=1
    li=d.values()
    for i in d.items():
        if i[1] == max(li):
            print(i[0] ,":",i[1])
    return 
HighestNoOfApps(incomedf)
        
```

```python
#Category with lowest number of apps
incomedf = readCSVdata(filepath)
def LowestNoOfApps(fp):
    l=[]
    d={}
    li=[]
    for i in range(len(fp.values[0::])):
        column = fp.values[i][12]
        l.append(column)
        if l[i] not in d:
            d[l[i]]=1
        else:
            d[l[i]]+=1
    li=d.values()
    for i in d.items():
        if i[1] == min(li):
            print(i[0] ,":",i[1])
    return 
LowestNoOfApps(incomedf)
```

```python
#Category with highest user rating
incomedf = readCSVdata(filepath)
def HighestUserRating(fp):
    l=[]
    d={}
    li=[]
    for i in range(len(fp.values[0::])):
        column = fp.values[i][8]
        l.append(column)
        if i not in d:
            d[fp.values[i][12]]=l[i]    
    li=d.values()
    for i in d.items():
        if i[1] == max(li):
            print(i[0],":",i[1])
    return
HighestUserRating(incomedf)
```

```python
# Average user rating for free apps
incomedf = readCSVdata(filepath)
l=[]

def FreeApps(fp):
    c=0
    for i in range(len(fp.values[0::])):
        column = fp.values[i][5]
        if column == 0.00:
            l.append(fp.values[i][8])
            c+=1
    print(sum(l))        
    return sum(l)/c
FreeApps(incomedf)
```

```python
# Average user rating for paid apps
incomedf = readCSVdata(filepath)
l=[]

def FreeApps(fp):
    c=0
    for i in range(len(fp.values[0::])):
        column = fp.values[i][5]
        if column > 0.00:
            l.append(fp.values[i][8])
            c+=1
    #print(sum(l))      
    return sum(l)/c
FreeApps(incomedf)
```
